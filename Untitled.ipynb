```python
import time
import psutil
%matplotlib notebook
import matplotlib.pyplot as plt
import IPython
import ipympl
```

```python
plt.style.use('ggplot')
# %matplotlib widget
# plt.rcParams['animation.html'] = 'jshtml'
```

```python
fig = plt.figure();
ax = fig.add_subplot(111);
```

```python
i = 0

x,y = [],[]

while True:
    x.append(i)
    y.append(psutil.cpu_percent())
    ax.plot(x,y,color='b')
    fig.canvas.draw()
    time.sleep(1)
    i += 1
```
